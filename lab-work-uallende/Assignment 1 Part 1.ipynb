7. A bag of words counts every occurrence of each word in a document, it is often used in document classification. Create a bag-of-words representation of this text

*e.g.* In the sentence "To be or not to be" the bag of words representation is
"TO": 2
"BE": 2,
"OR": 1,
"NOT": 1

```python
#7 bag of words

import operator
paragraphs = []

with open('./UlyssesChapter18.txt') as f:
    paragraphs = f.readlines()

words = "".join(paragraphs).split()
wordfreq = {}

for word in words:
    if word not in wordfreq.keys():
        wordfreq[word] = 1
    else:
        wordfreq[word] += 1

wordfreq = sorted(wordfreq.items(), key = operator.itemgetter(1), reverse = True)
print(wordfreq)
```
